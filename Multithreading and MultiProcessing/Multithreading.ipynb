### A Process is an instance of a computer Program
### Multiple Threads running on a processor/core >> MultiThreading

```python
import time
start = time.perf_counter()  # time.per_counter has higher precision than time.time() so here we hve used this

def test_func():
    print("Working")
    print("Sleep for 1 sec")
    time.sleep(1)
    print("Sleeping Completed")

test_func()
test_func()
test_func()
end = time.perf_counter()

print(f"The Program finshes in {round(end-start, 2)} seconds")
```

```python
import time
import threading
start = time.perf_counter()   # This is the main thread

def test_func():
    print("Working")
    print("Sleep for 1 sec")
    time.sleep(1)
    print("Sleeping Completed")

t1 = threading.Thread(target = test_func)  # Here test_func is not executed just passed
t2 = threading.Thread(target = test_func)  # Here test_func is not executed just passed

t1.start()  # Starting Thread 1 
t2.start()  # Starting Thread 2

t1.join()
t2.join()

end = time.perf_counter()

print(f"The Program finshes in {round(end-start, 2)} seconds")

# Here both threads ran simultaneously during sleep period
```

Main Thread
│
├── create t1
├── create t2
│
├── start t1 ──► test_func() ── sleep(1) ──► finish
│
├── start t2 ──► test_func() ── sleep(1) ──► finish
│
├── join t1 (wait)
├── join t2 (wait)
│
└── print execution time

** Threads in Python run concurrently not parallely meaning only one thread can run at a time in python

```python
import time
import threading
start = time.perf_counter()   # This is the main thread

def test_func():
    print("Working")
    print("Sleep for 1 sec")
    time.sleep(1)
    print("Sleeping Completed")

threads = []  # This list will contain all the threads
for i in range(10):  # 
    t = threading.Thread(target = test_func)
    t.start()  # Start the thread
    threads.append(t)  # Add this thread to threads list

for thread in threads:
    thread.join()   # Now we have joined all threads(Main thread will not be completed until all threads are completed)

end = time.perf_counter()

print(f"The Program finshes in {round(end-start, 2)} seconds")
```

Here all threads are finished in 1 sec whereas if it is executed one after the another it would take 10 sec

```python
import time
import threading
start = time.perf_counter()   # This is the main thread

url_list = [
    'https://github.com/itsfoss/text-script-files/blob/master/agatha.txt',
    'https://github.com/itsfoss/text-script-files/blob/master/agatha_complete.txt',
    'https://github.com/itsfoss/text-script-files/blob/master/sample_log_file.txt',
    'https://github.com/itsfoss/text-script-files/blob/master/sherlock.txt'
]  

data_list = ["data1.txt", "data2.txt", "data3.txt", "data4.txt"]  # This is how the downloaded files will be saved as

import urllib.request  # This provides functions to interact with url's

def file_download(url,filename):   # This is the download function
    urllib.request.urlretrieve(url,filename)

threads = []    # This the list where all thread objects will be stored
for i in range(len(url_list)):  # Loop through the length of url_list
    t = threading.Thread(target = file_download, args = (url_list[i], data_list[i])) # Download 1st url into data1.txt and so on
    t.start()   # Start the Thread
    threads.append(t)  # Append threads noe after the another

for thread in threads:   # Now perform join
    thread.join()

end = time.perf_counter()

print(f"The Program finshes in {round(end-start, 2)} seconds")
```

├── Create Thread-1
├── Start Thread-1

├── Create Thread-2
├── Start Thread-2

├── Create Thread-3
├── Start Thread-3
│
├── Create Thread-4
├── Start Thread-4
│
├── Join Thread-1
├── Join Thread-2
├── Join Thread-3

```python
# Does same work as above but lines of code lessens
# Preferred modern way
import time
import concurrent.futures
start = time.perf_counter()   # This is the main thread

url_list = [
    'https://github.com/itsfoss/text-script-files/blob/master/agatha.txt',
    'https://github.com/itsfoss/text-script-files/blob/master/agatha_complete.txt',
    'https://github.com/itsfoss/text-script-files/blob/master/sample_log_file.txt',
    'https://github.com/itsfoss/text-script-files/blob/master/sherlock.txt'
]  

data_list = ["data1.txt", "data2.txt", "data3.txt", "data4.txt"]  # This is how the downloaded files will be saved as

import urllib.request  # This provides functions to interact with url's

def file_download(url,filename):   # This is the download function
    urllib.request.urlretrieve(url,filename)

with concurrent.futures.ThreadPoolExecutor() as executor:  # Does all threading and join on it's own
    executor.map(file_download,url_list,data_list)

end = time.perf_counter()

print(f"The Program finshes in {round(end-start, 2)} seconds")
```

```python
import time
start = time.perf_counter()   # This is the main thread

# shared_counter = shared variable across all threads
shared_counter = 0
counter_lock = threading.Lock() # Used to synchronize access to the shared_counter means at one time only one thread can modify shared_counter

def increment_shared_counter(x):
    global shared_counter     # Means access the global variable shared_counter and not create new local variable
    with counter_lock:       # Acts like exception handling
        shared_counter = shared_counter + 1   # Increment of shared_counter by 1(All threads increment it one after the another)
        print(f"Thread {x}: incremented shared counter to {shared_counter}")
        time.sleep(1)    # Thread locks and sleep for 1 sec then passes the lock to the next thread

threads = [threading.Thread(target = increment_shared_counter, args = (i,)) for i in range(1,6)]  # This creates 5 threads but does not start them yet

for thread in threads:  # Now the created thread Starts
    thread.start()

for thread in threads:  # Perform join on the threads
    thread.join()

end = time.perf_counter()

print(f"The Program finshes in {round(end-start, 2)} seconds")
```
