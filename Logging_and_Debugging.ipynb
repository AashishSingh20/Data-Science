# Logging >> Keeps Track of events in a software run

```python
# Logging supports different levels of logging which helps tp categorize messages based on their severity
import logging
```

```python
# force = True is a parameter that is used to remove exisiting logging config and apply new one
logging.basicConfig(filename = "test.log", level = logging.INFO, force = True) 
```

```python
logging.info("This is a normal Info about the software")
logging.error("This is the error message")
logging.warning("This is a warning")
logging.debug("This is a debugging message")   # Debug ko agar info ke pehle likhte toh print hota but hamne badme likha hai isliye nahi print hua
logging.critical("This is a critical message")
logging.shutdown()
```

## Logging Heirarchy:
        1. DEBUG  >> Detailed debugging information(Lowest Level)
        2. INFO  >> General Information
        3. WARNING  >>  Warning Messages or potential issues
        4. ERROR  >>  Errors
        5. CRITICAL  >>  Serious Errors


```python
# Create a file named test_new.log with lowest level(debug) and format to print in the log is asc to time then messages then force this logging
logging.basicConfig(filename = "test_new.log", level = logging.DEBUG, format = '%(asctime)s %(message)s', force = True) 
```

```python
logging.debug("This is a debugging message")
logging.info("This is General info message")
logging.warning("This is warning message")
logging.shutdown()
```

```python
import logging
logging.basicConfig(filename = "program.log", level = logging.DEBUG, format = '%(asctime)s %(levelname)s %(message)s', force = True) 
```

```python
# Write a prog to seperate the integer and string in 2 lists seperately
l = [1,"hello", [2,"hi"], 5, ["python", 8]]

l1_int = []  # This list will store integers 
l2_str = []  # This list will store string 

for i in l:  # Traverse through the main list l
    logging.info(f"Processing each elements {i}")  # Process each elements one by one 

    if type(i) == list:  # If the element is a list then 
        for j in i:   # Go deeper and go more deep
            logging.info(f"Processing SubList element: {j}")  # Process sublist elements
            
            if type(j) == int:  # if type of the sublist element is integer then add it to list 1
                l1_int.append(j)

    elif type(i) == int:    # if type of list element is integer then add it to list l1
        l1_int.append(i)
    else:   # Else if element of list or sublist is string then add it to list l2
        l2_str.append(i)

logging.info(f"The final result is: {l1_int}, {l2_str}")   # Print final result
logging.shutdown()
```
