```python
class Student:
    def __init__(self):
        self.phone = 987654
    
    def __str__(self):  # __str__ will represent a string representation of  
        return "This method overrides the print statement of object method"
```

```python
# repr

class MyClass:
    def __init__(self, x):
        self.x = x
    
    def __repr__(self):
        return f"MyClass({self.x})"
```

```python
obj = MyClass(5)
```

```python
repr(obj)  # It will return strng representation of object
```

```python
print(repr(obj))
```

```python
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y
```

```python
p1 = Point(2,3)
p2 = Point(1,3)
```

```python
print(p1 == p2)
```
