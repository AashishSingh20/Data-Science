```python
class Student:
    def __init__(self,name,price):
        self.__name = name
        self.__price = price
    
    @property   # This will make access_price as a attribute of the class 
    def access_price(self):
        return self.__price
    
    @access_price.setter  # Helps us to set new_price
    def price_set(self,price_new):
        self.__price = price_new

    @access_price.deleter
    def del_price(self):
        del self.__price
```

```python
stud = Student("Ritik", 2000)
stud.access_price
```

```python
stud.price_set = 1500
stud.access_price
```

```python
del stud.del_price
```

```python
stud.access_price  # Will throw error as we have already deleted price
```

```python
class Circle:
    def __init__(self,radius):
        self.__radius = radius

    @property
    def radius(self):
        return self.__radius
    
    @radius.setter
    def radius(self,value):
        if(value <= 0):
            raise ValueError("Incorrect Radius")
        self.__radius = value
        
    def area(self):
        return 3.14 * self.__radius * self.__radius 
```

```python
c = Circle(3)
c.radius
```

```python
c.radius = 10
c.radius
```

```python
c.radius = -4
c.radius
```

```python
c.area()
```
