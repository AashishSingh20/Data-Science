## Method Overloading  (Same method taking different forms inside a class) ##
[Last Method Overrides the previous method]

```python
class Student:
    def student(self):
        print("Welcome")
    def student(self,name = ""):
        print("Welcome",name)
    def student(self,name = "", course = ""):
        print("Welcome",name,course)
```

```python
stu = Student() 
stu.student()
```

```python
stu.student("Aashish")
```

```python
stu.student("Aashish","Python")
```

## Public  (Accessible anywhere from inside or outside the class)

```python
class Student:
    def __init__(self,name,degree):
        self.name = name
        self.degree = degree
```

```python
stud = Student("Aashish", "Bachelors")
```

```python
stud.name
```

```python
stud.degree
```

```python
stud.degree = "Masters"
stud.degree
```

## Private  (Only Accessible within the class)

```python
class Student:
    def __init__(self,name,degree):
        self.name = name
        self.__degree = degree  # Using __ to make methods private

    def show(self):   # This gives the liberty to the user to see private members but they cannot edit it
        # Accessing the public and private data member
        print("Name:", self.name, "Degree:", self.__degree)
```

```python
stud = Student("Aashish", "Bachelors")
```

```python
stud.name
# stud.__degree
```

```python
stud._Student__degree = "Masters" # We can Access Private members like this(Only accessible when we know class name and data member to edit)
stud._Student__degree     # We can also change values in private members using getters and setters
```

```python
stud.show()
```

```python
class Bank:
    def __init__(self,balance):
        self.__balance = balance
    def deposit(self, amount):
        self.__balance = self.__balance + amount
    def withdraw(self, amount):
        if self.__balance >= amount:
            self.__balance = self.__balance - amount
            return True
        else:
            return False
    
    def get_balance(self):
        return self.__balance
```

```python
acc = Bank(1000)
acc.get_balance()
```

```python
acc.deposit(200)
acc.get_balance()
```

```python
acc.withdraw(400)
acc.get_balance()
```

```python
acc.withdraw(4000)
```

## Protected (Can be accessed within the class within the class and it's derived class) [_]

```python
class College:
    def __init__(self):
        self._college_name = "Raisoni"
    
class Student(College):
    def __init__(self,name):
        self.name = name
        # College.__init__(self)  # Syntax = classname.function(you want to access)
        super().__init__()  # Syntax = classname.function(you want to access)  # Same as Above
    
    def show(self):
        print("Name:", self.name, "College:", self._college_name)
```

```python
stud = Student("Aashish")
```

```python
stud.show()
```

```python
coll = College()
coll._college_name  # We can also access private members using this 
```
