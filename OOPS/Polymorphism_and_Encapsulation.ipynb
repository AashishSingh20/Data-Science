```python
class Teacher:
    def lec_info(self):
        print("This is Teacher's Lecture")
```

```python
class Student:
    def lec_info(self):
        print("This is Student's Lecture")
```

```python
def parcer(class_obj):
    for i in class_obj:
        i.lec_info()
```

```python
obj1 = Teacher()
obj2 = Student()
```

```python
class_obj = [obj1,obj2]
```

```python
parcer(class_obj)
```
