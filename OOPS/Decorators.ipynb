# Decoraters # (Allows to modify/extend the existing behaviour of functions/ class)

```python
def my_deco_func():
    print("Lines Before the Computation.")
    print(2+22)
    print("Lines After the Computation.")
```

```python
my_deco_func()
```

```python
def my_deco(func):  # Decorater function that takes another function as argument 
    def wrapper():
        print("Lines Before the Computation.")
        func()
        print("Lines After the Computation.")
    return wrapper
        
```

```python
@my_deco
def say_hello():
    print("Say Hello")
# When say_hello is called, it is actually calling wrapper() which in turn calls say_hello
```

```python
say_hello()
```

```python
import time
def timer_decorater(func):
    def timer():
        start = time.time()
        func()
        end = time.time()
        print(end-start)
    return timer
```

```python
@timer_decorater
def func_test():
    print(1000**200)
```

```python
func_test()
```

```python
# Class Decorator
class MyDecorator:
    def __init__(self,func):
        self.func = func
    def __call__(self): # Called when instance of a class(object) is used as a function
        print("Something is happening before func")
        self.func()
        print("Something is happening after func")
```

```python
@MyDecorator  # __call__ is a special method which is invoked when you call a decorator class instance(object)
def say_hello():
    print("Hello")
```

```python
say_hello()
```

# Built in decorators >> classmethod, staticMethod property

## Static Method >> Can be called without creating an instance of a class(object)

```python
class Math:
    def add(self,x,y):
        return x+y
```

```python
a = Math() #make object/instance
```

```python
a.add(3,4)  # This was a regular class
```

```python
class Math:
    @staticmethod 
    def add(x,y):
        return x+y
```

```python
Math.add(4,5)
```

## class method >> takes class itself as first argument  

```python
class Math:
    @classmethod  # Takes reference to the class itself to modify and access class level attributes
    def add(cls,x,y):
        return cls.__name__, x+y  # cls.__name__ >> Math
```

```python
Math.add(2,3)
```

## Property Decorators >> it allows methods to be accessed as attribute

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        radius = self.radius
        return 3.14 * radius ** 2
```

```python
c = Circle(4)
```

```python
c.area()  # Normal Call
```

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius
    @property            # Property Decorator
    def area(self):
        radius = self.radius
        return 3.14 * radius ** 2
```

```python
c = Circle(4)
```

```python
c.area  
```
