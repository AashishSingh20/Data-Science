# Class Methods # ( Bound to class, can access and modify methods and attributes of all instances)

```python
class Student:
    def __init__(self,name):  # Method of instance
        self.name = name
```

```python
obj = Student("Aashish")
obj.name
```

```python
class Student:
    total_students = 0
    def __init__(self,name):
        self.name = name
        Student.total_students = Student.total_students + 1
    
    @classmethod  # Bound to the class
    def student_details(cls,name):
        return cls(name)
```

```python
Student.total_students
```

```python
obj1 = Student.student_details("Aashish")
print(obj1.name)
Student.total_students  
```

```python
obj2 = Student.student_details("Kaushik")
Student.total_students
```

```python
# To make external class method to be part of the class
class Student:
    total_students = 0
    def __init__(self,name):
        self.name = name
        Student.total_students = Student.total_students + 1
    
    @classmethod  # Bound to the class
    def student_details(cls,name):
        return cls(name)
```

```python
def course_details(self,course_name):
    print("The details of the course is:", course_name)
```

```python
Student.course_details = classmethod(course_details)  # Adding the external class method 
```

```python
Student.course_details("Python")
```

```python
# Deleting a Class Method 
del(Student.course_details)
```

```python
Student.course_details("Python")
```

```python
obj3 = Student.student_details("Raj")
print(obj3.name)
Student.total_students  
```

```python
delattr(Student, "total_students")
```

```python
obj3 = Student.student_details("Raj")
print(obj3.name)
Student.total_students  
```

# Static Methods # (It doesn't have access of the class and it cannot modify class as well)
 It works with the paramaters that is passed 

```python
class Calculator:
    
    @staticmethod
    def add(a,b):
        return a+b
    def subtract(a,b):
        return a-b
```

```python
Calculator.add(4,1)
```

```python
Calculator.subtract(6,8)
```
