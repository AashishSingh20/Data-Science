#  Inheritance  #

```python
class BaseClass:
    # Body of Base Class
    pass 
class DerivedClass(BaseClass):
    # Body of Derived Class
    pass
```

## Single Inheritance(When a class has only one parent class) ##

```python
class Father:
    def father_property(self):
        print("This is Father's Class")
```

```python
class Son(Father):
    def inherit(self):
        print("This is Son's class")
```

```python
child_obj = Son()
child_obj.father_property()
child_obj.inherit()
```

```python
class Fruit:
    def fruit_info(self):
        print("This is Parent Class")
        
class Apple(Fruit):
    def fruit_info(self):
        print("This is Child class")
    def apple_info(self):
        print("This is also in Child Class")
```

```python
apple = Apple()
apple.fruit_info()  # Here function of parent class is overridden by function in child class this is method overriding
apple.apple_info()
```

## Multilevel Inheritance ##

```python
class GrandFather():
    def property_info(self):
        print("This is Grandfather's Property")

class Father(GrandFather):
    def property_info2(self):
        print("This is Father's Property + Grandfather's ")

class Son(Father):
    def property_info3(self):
        print("This is Son's Property + Father's + GrandFather's")
```

```python
property = Son()
property.property_info()
property.property_info2()
property.property_info3()
```

## Mutliple Inheritance (One Child class may inherit from multiple Parent Class) ##

```python
class ParentClass1:
    def method1(self):
        print("Method 1 of Parent class 1")
    
class ParentClass2:
    def method2(self):
        print("Method 2 of Parent class 2")
        
class childClass(ParentClass1,ParentClass2):
    def method(self):
        print("Child Method")

    
```

```python
child = childClass()
child.method1()
child.method2()
child.method()
```

```python
class A:
    def method(self):
        print("Method of Class A")
    
class B(A):
    def method(self):
        print("Method of Class B")

class C(A):
    def method(self):
        print("Method of Class C")

class D(C,B):
    pass
```

```python
child = D()  # Diamond Problem
child.method()   # Joh class pehle likha hai D ke argument mein woh print hoga
```

## Heirarchical Inheritance ##

```python
class Vehicle:
    def vehicle_info(self):
        print("This is Vehicle Class")
    
class Car(Vehicle):
    def car_info(self):
        print("This is Car Class")

class Bus(Vehicle):
    def bus_info(self):
        print("This is Bus Class")

class Truck(Vehicle):
    def truck_info(self):
        print("This is Truck Class")
```

```python
car = Car()
car.vehicle_info()
car.car_info()

bus = Bus()
bus.vehicle_info()
bus.bus_info()

truck = Truck()
truck.vehicle_info()
truck.truck_info()
```

## Hybrid Inheritance ##

```python
class Vehicle:
    def vehicle_info(self):
        print("This is Vehicle Class")
    
class Car(Vehicle):
    def car_info(self):
        print("This is Car Class")

class Bus(Vehicle):
    def bus_info(self):
        print("This is Bus Class")

class SportsCar(Car,Vehicle):
    def sports_car_info(self):
        print("This is Sports Car Class")
```

```python
s_car = SportsCar()
s_car.vehicle_info()
s_car.car_info()
s_car.sports_car_info()
```

# Abstraction (Concept of hiding Complete Implementation) #

```python
import abc

class College:
    @abc.abstractmethod
    def student_Details(self):
        pass

    @abc.abstractmethod
    def student_Assignment(self):
        pass

    @abc.abstractmethod
    def student_Marks(self):
        pass
```

```python
class Python(College):
    def student_Details(self):
        return "Python Class Student Details"

    def student_Marks(self):
        return "Python Class Student Marks"

    def student_Assignment(self):
        return "Python Class Student Assignment"
```

```python
class WebDev(College):
    def student_Details(self):
        return "Web Dev Class Student Details"

    def student_Marks(self):
        return "Web Dev Class Student Marks"

    def student_Assignment(self):
        return "Web Dev Class Student Assignment"
```

```python
py = Python()
py.student_Details()
```

```python
wb = WebDev()
wb.student_Marks()
```
