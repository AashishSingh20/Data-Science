## r -> Read Mode
## w -> Write Mode
## a -> Append Mode 
## r+ -> Both Reading and Writing

```python
file = open("file.txt", "w")  # Open file in write mode after truncating the file first(If file does not exist then creates the file)
```

```python
file.write("Helo, I am Aashish")  # Here the number in output signifies the no of bytes we have written
```

```python
file.close()  # This updates the new changes in the file and closes it 
```

```python
file = open("file.txt", "w")
file.write("This is First Line\n")
file.write("This is Second Line\n")
file.write("This is Third Line\n")
file.write("This is Fourth Line\n")
file.close()
```

```python
file = open("file.txt", "a")  # Appends the file 
file.write("This is my Fifth Line")
file.close()
```

```python
file = open("file.txt","r")

for i in file:
    print(i)
    
file.write("Hello")   # Not possible as file is opened in read mode
```
