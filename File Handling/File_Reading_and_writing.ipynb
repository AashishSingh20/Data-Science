### Csv >> Comma Seperated Values

```python
data = [["Name", "Course", "Fee"],
        ["Aashish", "Python", "500"],
        ["Karan", "ML", "2500"]]
```

```python
data 
```

```python
for i in data:
    print(i)
```

```python
import csv

with open("test_csv.csv", "w", newline="") as f:  # newline is written to prevent from newline being written
    w = csv.writer(f)
    for i in data :
        w.writerow(i)
```

```python
with open("test_csv.csv", "r") as f:
    r = csv.reader(f)
    for i in r:
        print(i)
```

### Buffered Reading and Writing (It allows the program to read and write the data in chunks)

```python
# Large amounts of data cannot be loaded at once as the system can crash so we use buffer to read or write data in small chunks
# Python already used buffering but we use buffer when we want explicit control over it

import io
with open("test_buf.txt", "wb") as f:  
    file = io.BufferedWriter(f)
    file.write(b"Machine learning (ML) is a field of study in artificial intelligence concerned with the development and study of statistical algorithms that can learn from data and generalize to unseen data, and thus perform tasks without being explicitly programmed.[1] Advances in the field of deep learning have allowed neural networks, a class of statistical algorithms, to surpass many previous machine learning approaches in performance.")
    file.write(b"I am a Boy")
    file.write(b"I am a CS Student")
    file.flush()
```

```python
with open("test_buf.txt", "rb") as f:
    file = io.BufferedReader(f)
    data = file.read(100)  # 100 bytes ke chunk mein data read karega
    print(data)
```
