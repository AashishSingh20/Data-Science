## Custom Exception

```python
income = int(input("Enter your income: "))
print(f"My income is: {income}")
```

```python
class ValidateSalary(Exception):  # Will contain all exceptions
    def __init__(self, msg):
        self.msg = msg
```

```python
def Validate_Salary(salary):
    if salary <= 0:
        raise ValidateSalary("Salary cannot be negative or 0")
    elif salary > 100000000000:
        raise ValidateSalary("This high salary cannot be possible")
    else:
        print("Salary is Valid")
    
```

```python
try:
    salary = int(input("Enter your Salary: "))
    Validate_Salary(salary)
except ValidateSalary as e:
    print(e)
```

### Best Practices:
1) Always use specific exception
2) Always Provide clear message for errors
3) Always try logging
4) Don't write unnecessary exceptions
5) Give Proper Documentation
6) Always cleanup/free used resources

```python
try:
    int("hello")
except ValueError as e:
    print("The string cannot be converted into an integer", e)
```

```python
try:
    int("hello")
except (ValueError, TypeError) as e:  # Used when we want to use 2 types of error
    print("The string cannot be converted into an integer", e)
```

```python
try:
    f = open("example1.txt", "r")
    f.read()

except Exception as e:  # This will be executed an below exception will not as Exception is the superclass of all exceptions
    print("File not Found!")
except FileNotFoundError as e:
    print("File is not Found")
```

```python
import logging
logging.basicConfig(filename = "program.log", level = logging.DEBUG, force = True)
try:
    f = open("example1.txt", "w")
    f.write("Hello")

except FileNotFoundError as e:
    logging.error(f"File not Found {e}")
finally:
    f.close()
```
