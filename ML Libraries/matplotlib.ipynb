## Line Plot
#### Use when time component is involved on x axis
#### Depict the relation between 2 variables in the trend

```python
x = [1,2,3,4,5]
y = [6,7,3,9,4]

plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.title("Line Plot the understand how data visualization works")   # Gives title to the graph
plt.xlabel("x Axis")  # Gives label to x axis
plt.ylabel("y Axis")  # Gives label to y axis
plt.show() 
```

```python
plt.plot(x,y, color = "teal")
plt.title("Line Plot the understand how data visualization works")
plt.xlabel("x Axis")
plt.ylabel("y Axis")
plt.show()
```

```python
plt.plot(x,y, color = "teal", marker = "o")
plt.title("Line Plot the understand how data visualization works")
plt.xlabel("x Axis")
plt.ylabel("y Axis")
plt.show()
```

```python
plt.plot(x,y, color = "teal", marker = "o", linestyle = "--")
plt.title("Line Plot the understand how data visualization works")
plt.xlabel("x Axis")
plt.ylabel("y Axis")
plt.show()
```

```python
plt.plot(x,y, color = "teal", marker = "o", linestyle = "--", linewidth = "3")
plt.title("Line Plot the understand how data visualization works")
plt.xlabel("x Axis")
plt.ylabel("y Axis")
plt.show()
```

```python
plt.plot(x,y, color = "teal", marker = "o", linestyle = "--", linewidth = "3", markersize = 10)
plt.title("Line Plot the understand how data visualization works")
plt.xlabel("x Axis")
plt.ylabel("y Axis")
plt.show()
```

```python
plt.plot(x,y, color = "teal", marker = "o", linestyle = "--", linewidth = "3", markersize = 10)
plt.title("Line Plot the understand how data visualization works")
plt.xlabel("x Axis")
plt.ylabel("y Axis")
plt.grid()
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [3,6,1,3,7]
y2 = [5,2,1,4,3]

plt.plot(x,y)
```

```python
plt.plot(x,y1, color = "teal", marker = "o", linestyle = "--", linewidth = "3", markersize = 10, label = "line_y1")
plt.plot(x,y2, color = "coral", marker = "o", linestyle = "--", linewidth = "3", markersize = 10, label = "line_y2")

plt.xlabel("x data axis")
plt.ylabel("y data axis")

plt.title("Multiple lines in one plot")
plt.legend()   # Lines ke upar ka label show hone lagega

plt.show()
```

```python
x = np.linspace(0,10,100)  # Will generate 100 data points between 0 and 10
x
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.sin(x), '.-')
plt.plot(x, np.cos(x), '.-')
```

```python
# Creating Subplots in a plot
plt.Figure()   # Creates a page 
plt.subplot(1,2,1)  # Divide page into 1 row, 2 column then 1st part mein sin(x) ka graph plot karo
plt.plot(x, np.sin(x))

plt.subplot(1,2,2)  # 1 row and 2 column wale page mein 2nd column wale part mein cons(x) ka graph plot karo
plt.plot(x, np.cos(x))  

plt.show()
```

```python
plt.Figure()   # Creates a page 
plt.subplot(2,2,1)  # Divide page into 2 rows, 2 column then 1st part mein sin(x) ka graph plot karo
plt.plot(x, np.sin(x))

plt.subplot(2,2,2)  # 2 row and 2 column wale page mein 2nd column wale part mein cons(x) ka graph plot karo
plt.plot(x, np.cos(x))  

plt.subplot(2,2,3)  # 2 row and 2 column wale page mein 3rd column wale part mein tan(x) ka graph plot karo
plt.plot(x, np.tan(x)) 

plt.subplot(2,2,4)  # 2 row and 2 column wale page mein 4th column wale part mein sec(x) ka graph plot karo
plt.plot(x, np.exp(x)) 

plt.show()
```

```python
plt.Figure()

plt.subplot(2,1,1) # Creates 2 rows and 1 column then create sin(x) plot on 1st part
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)  # Creates 2 rows and 1 column then create cos(x) plot on 1st part
plt.plot(x, np.cos(x))
```

## Histogram
#### Shows how continuous or numerical data is distributed

```python
data = np.random.randn(1000)   # 10 bins
plt.hist(data, color = 'turquoise')
plt.show()
```

```python
data = np.random.randn(1000)   # 10 bins
plt.hist(data, color = 'green', bins = 20)
plt.show()
```

```python
# Stacked Histogram

data1 = np.random.normal(0,1,1000)
data2 = np.random.normal(0,1,1000)
```

```python
plt.hist([data1,data2], bins = 20, color = ["black","brown"], label = ["dist1","dist2"])
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Stacked Histogram")

plt.show()
```

## Bar Chart 
#### Used to understand the categorical variables and it is not continuous like histogram

```python
a = ['a','b','c','d']
b = [4,1,6,2]

plt.bar(a,b, color = "brown")
```

```python
a = ['a','b','c','d']
b = [4,1,6,2]

plt.barh(a,b, color = "brown")  # barh gets us vertical bar chart
```

## PieChart
#### Used to understand the percent distribution of data

```python
langs = ['C', 'C++', 'Java', "Python"]
students = [10,40,55,25]

plt.pie(students, labels = langs, autopct = '%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
explode = (0.1, 0.1, 0.1, 0.1)  # Will seperate all part of pie chart

langs = ['C', 'C++', 'Java', "Python"]
students = [10,40,55,25]

plt.pie(students, labels = langs, autopct = '%1.1f%%', explode = explode, shadow = True)
plt.title("Pie Chart")
plt.show()
```
