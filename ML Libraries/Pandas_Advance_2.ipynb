### Imputation of missing values
#### Numeric data >> replace with mean and median  >> outlier treatment done >> then mean else median
#### Categorical data  >> replace with mode
#### Or impute missing values with some constants like 0

```python
df1
```

```python
df1.fillna("somevalue")   # Fills somevalue in place if null
```

```python
df1.fillna(0)   # Fills 0 in place if null
```

```python
df1.fillna(df1.mean())   # Fills somevalue in place if null
```

```python
data1 = {
    'A' : [1,2,3,None,4,None,5,6,None,7,8],
    'B' : [1,None,None,2,3,4,5,None,6,7,None]
}

df2 = pd.DataFrame(data1)
df2
```

```python
df2.A.fillna(df2['A'].mean())  # Impute null values with mean 
```

```python
df2.A.fillna(df2['A'].median())  # Impute null values with median 
```

```python
df2
```

```python
# df2.fillna(method = 'ffill')   # This is forward fill pehle wali(non-null) value ko null ki jagah impute kardo
df2.ffill()  # New syntax
```

```python
df2
```

```python
df2.fillna(method = 'bfill')   # This is backward fill baad wali(non-null) value ko null ki jagah impute kardo
df2.bfill()  # New Syntax
```

```python
df = pd.read_csv('titanic dataset.csv')
df
```

```python
df.mean(numeric_only = True)
```

```python
df.median(numeric_only = True)
```

```python
df.std(numeric_only = True)
```

```python
# What is the average fare paid by people who survived?
```

```python
df[df.Survived == 1]['Fare'].mean()  # Average Fare paid by people who survived
```

```python
df[df.Survived == 0]['Fare'].mean()  # Average Fare paid by people who didn't survive
```

```python
# groupby

df.groupby('Survived').mean(numeric_only = True)  # Gives average of all attibutes in survived class
```

```python
import numpy as np
df.groupby('Survived')['Fare'].agg(['min','max','mean','median','count','std','var'])
```

```python
df.groupby(['Sex','Pclass'])['Survived'].sum()
```

```python
df.groupby(['Sex','Pclass'])['Survived'].sum().to_frame()
```

```python
df.groupby(['Sex','Pclass'])['Survived'].sum().unstack()
```

```python
a = df.groupby('Pclass').sum(numeric_only = True)  # Sum of values of all Values by Pclass class
a
```
