### Merge
Merging on the basis of common elements

```python
df1 = pd.DataFrame({'key1': [1,2,4,5,6],
                    'key2': [4,5,6,7,8],
                    'key3': [3,4,5,6,6]
                   })

df1
```

```python
df2 = pd.DataFrame({'key1': [1,2,45,6,65],
                    'key4': [56,5,6,7,8],
                    'key5': [3,56,5,6,6]
                   })

df2
```

```python
pd.merge(df1,df2, how = 'inner')  # Key1 ke common values jis row mein hai un rows ke elements ko new tale mein add kardo
```

```python
pd.merge(df1,df2, how = 'left')  # Left join >> Left ke saare elements and key1 ke common elements jis row mein hai df2 ke andar joh aayenge
```

```python
pd.merge(df1,df2, how = 'right')   # Right join >> Right ke saare elements and key1 ke common elements jis row mein hai df1 ke andar joh aayenge
```

```python
pd.merge(df1,df2, how = 'cross')   # 1st row of df1 with all rows of df2, then 2nd row of df1 with all rows of df2 and so on
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2, how = 'outer')  # Will give all keys from both dataframes (key1/common column ke saare values aajayenge)
```

```python
pd.merge(df1,df2, how = 'left', left_on = 'key2', right_on = 'key4')  # Custom comman key  
```

### Join 
Merging on the basis of indexes

```python
df1 = pd.DataFrame({'key1': [1,2,4,5,6],
                    'key2': [4,5,6,7,8],
                    'key3': [3,4,5,6,6]
                   },
index = ['a','b','c','d','e']
)

df1
```

```python
df2 = pd.DataFrame({'key4': [1,2,45,6,65],
                    'key5': [56,5,6,7,8],
                    'key6': [3,56,5,6,6]
                   },
index = ['a','b','h','i','j']
)

df2
```

```python
df1.join(df2, how = 'inner')  # Join at common index of df1 and df2
```

```python
df1.join(df2, how = 'left')
```

```python
df1.join(df2, how = 'right')
```

```python
df1.join(df2, how = 'outer')
```

```python
df1.join(df2, how = 'cross')
```
