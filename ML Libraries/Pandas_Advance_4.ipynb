```python
df2 = pd.DataFrame({'a' : [1,2,3,4,5,6,7,8]})
df2
```

```python
df2.rolling(window = 1).mean()  # Provides Rolling Window Calculations.
```

```python
df2.rolling(window = 2).mean()  # Gives mean from 2 window size like mean of 1,2 then mean of 2,3 then 3,4 and so on
# Since 1 ke pehle koi element nahi hai isliye NaN show ho raha hai
```

```python
df3 = df2.rolling(window = 3).mean()
df3
```

```python
df3.dropna(inplace = True)
df3
```

```python
# Converting the above values from float to integer

df3['a_int'] = df3['a'].astype(int)
df3
```

```python
df2['Group'] = np.random.choice(['A','B'], size = 8)  # Assigns Random groups to all values from column a
df2
```

```python
df2.groupby('Group')['a'].rolling(window = 2).sum()  # Group ke according group banayega and then perform rolling sum on window size = 2
```
