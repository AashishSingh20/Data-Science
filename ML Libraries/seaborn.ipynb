```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
df
```

```python
# Numerical Data Plotting

sns.relplot(x = 'total_bill', y = 'tip', data = df)  # relplot helps us to understand the relationship between 2 variables(defaukt : scatter plot)
```

```python
sns.scatterplot(x = df.total_bill, y = df.tip, data = df, hue = 'smoker')
```

###### Insights:-
1) As Bill Increases the tipping amount also increases
2) Highest tip was paid by a smoker

```python
sns.scatterplot(x = df.total_bill, y = df.tip, data = df, hue = 'size')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, style = 'smoker', hue = 'time')
```

```python
val = np.random.randn(500).cumsum()
time = np.arange(500)

df = pd.DataFrame({'time': time, 'val': val,})
df
```

```python
# Using relplot to make a line plot

sns.relplot(x = 'time', y = 'val', kind = 'line', data = df, sort = True)
```
