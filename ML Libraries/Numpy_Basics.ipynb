```python
np.matrix([1,2,3]) # A Specialized Array, 2D
```

```python
lis = [1,2,3,4]
np.asarray(lis)   # Converts input to an array
```

```python
np.matrix([1,2,3])
np.asanyarray()  # Will convert only those inputs which are not an array to an array(Will not convert subclasses of array to array like matrix)
```

```python
np.fromfunction(lambda i, j: i==j, (2,2))  # This generates an array from the function
```

```python
np.fromfunction(lambda i, j: i*j, (3,3))  # This generates an array from the function
```

```python
iterable = (i for i in range(5))
```

```python
np.fromiter(iterable, int)   # Creates 1D Array from iterable object
```

```python
np.fromstring("10 12 14", sep = " ")  # This method only works for numerical values
```

```python
np.arange(1,10,0.2)  # Gives values from 1 to 10 with 0.2 difference
```

```python
np.linspace(1,5,10)  # This means give me 10 numbers between 1 to 5 which are evenly spaced
```

```python
np.logspace(1,5,10)  # Return numbers between 1 to 5 which are evenly spaced on a logscale[Base(default) = 10]
```

```python
np.logspace(1,5,10, base = 2)
```
